## Coeficientes de Fresnel

### Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Funciones 
Describen la Ley de Snell y las ecuaciones de los Coeficientes de Fresnel

```python
def Snell(Thetai, ni, nt):
    return np.arcsin(ni*np.sin(Thetai)/nt)
    
def rvals(Thetai, Thetat, ni, nt):
    rperpendicular = (ni*np.cos(Thetai)-nt*np.cos(Thetat))/(ni*np.cos(Thetai)+nt*np.cos(Thetat))
    rparallel = (nt*np.cos(Thetai)-ni*np.cos(Thetat))/(nt*np.cos(Thetai)+ni*np.cos(Thetat))
    return np.asarray([rperpendicular, rparallel])

def tvals(Thetai, Thetat, ni, nt):
    tperpendicular = (2*ni*np.cos(Thetai))/(ni*np.cos(Thetai)+nt*np.cos(Thetat))
    tparallel = (2*ni*np.cos(Thetai))/(nt*np.cos(Thetai)+ni*np.cos(Thetat))
    return np.asarray([tperpendicular, tparallel])
```

### Caso de Estudio - Interfaz Agua-Aire

```python
#Refractive Index
ni = 1. #Air
nt = 1.33158 #Water

#Angles
Thetai = np.linspace(0.,np.pi/2,1000)
Thetat = Snell(Thetai, ni, nt)

#Fresnel Coefficients
r = rvals(Thetai, Thetat, ni, nt)
t = tvals(Thetai, Thetat, ni, nt)

#Graphic
plt.plot(Thetai*180./np.pi, t[1,:], c='green', label = r'$t_{\parallel}$')
plt.plot(Thetai*180./np.pi, t[0,:], c='blue', label = r'$t_{\perp}$')
plt.plot(Thetai*180./np.pi, r[1,:], c='orange', label = r'$r_{\parallel}$')
plt.plot(Thetai*180./np.pi, r[0,:], c='red', label = r'$r_{\perp}$')
plt.title('Coeficientes de Fresnel para una Interfaz de Agua y Aire')
plt.xlabel('Ángulo de Incidencia')
plt.ylabel('Coeficientes de Fresnel')
plt.legend()
plt.grid()
plt.savefig('FresnelCoefficients.pdf')
plt.plot()
```
